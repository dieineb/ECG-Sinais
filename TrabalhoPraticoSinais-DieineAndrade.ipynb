```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import fftpack
```

```python
opentxt = open('ECG.txt')#importação do sinal original
ecg = np.loadtxt(opentxt)
N = 600
Ts = 1.0 /600 
t = np.arange (0,len(ecg))*Ts
plt.figure(figsize=(10,5))
plt.plot (t, ecg)
plt.title("Sinal ECG Original Completo")
plt.xlabel("Tempo(s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show
```

```python
#(a)plotar o sinal no domínio do TEMPO,ou seja, amplitude da amostra  x segundo, Janela mínima de 10s de sinal
pos_inicial = 0
segundos = 10
pontos_amostra = segundos*600
pontos = segundos/pontos_amostra
t = np.arange(0, segundos, pontos)
# Corte dos ruídos inicial e final
cut_ecg = ecg[45000: 460000]#Verifiquei valores até observar amostra sem ruídos para o corte
sinal_cortado = cut_ecg[pos_inicial:pontos_amostra+pos_inicial]
plt.figure(figsize=(10,5))
plt.title("ECG Sinal Ampliado 10s")
plt.xlabel("Tempo(s)")
plt.ylabel("Amplitude")
plt.grid()
plt.plot(t, sinal_cortado,"green")
```

```python
#Espectro da Amplitude do Sinal com remoção de Ruídos brutos iniciais

sinal_fft = fftpack.fft(sinal_cortado)
amplitude = np.abs(sinal_fft)
freq_amostra = fftpack.fftfreq(sinal_cortado.size, d=Ts)

plt.figure(figsize=(10,5))
plt.title("Espectro da Amplitude do Sinal Ampliado") 
plt.xlabel("Frequencia(Hz)")
plt.ylabel("Amplitude ")
plt.plot(freq_amostra,amplitude)
plt.grid()

#Somente freq positivas(ffp)
ffp= np.where(freq_amostra > 0)
freq_pos = freq_amostra[ffp]
freq_max = freq_pos[amplitude[ffp].argmax()]

#Verificação da maior amplitude e maior frequência Janela 10s
print("Maior amplitude:", amplitude.max())
print("Maior frequência:", freq_amostra.max()/100)

plt.show()
```

```python
#Espectro da amplitude do Sinal - Janela 10 segundos

plt.figure(figsize=(10,5))
plt.xlim(left=0, right=25)
plt.stem(freq_amostra, amplitude, 'green', use_line_collection= True)
plt.title("Espectro da Amplitude do Sinal - Janela 10s") 
plt.xlabel("Frequencia(Hz)")
plt.ylabel("Amplitude ")
plt.grid()

plt.show()
```
